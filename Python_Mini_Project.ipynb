## Finance Tracker

**Objective:** To find personal financial budget at the end of the month.

In our daily life, we all maintain our own personal expenses.At times, it is common that we lose track of our expenses and run out of budget. So, we need a way to keep up with our budget. To overcome this we can use our skills in Python to create a program that will allow us to keep track of our finance expenses. So, with the help of the program created, users can easily set a budget and add up their expenses to make better financial decisions. 

- Note: Amount entered beyond the budget will not be added up.



```python
def main():
    overallbudget = 10000
    selection=0
    while (selection != '8'):    #'6' string
        print("Enter 1 to add on expenses")
        print("Enter 2 to remove expenses")
        print("Enter 3 to add revenue")
        print("Enter 4 to subract revenue")
        print("Enter 5 to exit")

        selection = int(input("Enter your selection:"))
        
        if selection == 1: 
            overallbudget = add_expenses(overallbudget)
        elif selection == 2:
            overallbudget = subtract_expenses(overallbudget)
        elif selection == 3:
            overallbudget = add_revenue(overallbudget)
        elif selection == 4:
            overallbudget = subtract_revenue(overallbudget)
        elif selection == 5:
            print("Thank You!")
        else:
            print("Wrong entry, choose another option")


#Additon of expenses
def add_expenses(overallbudget):
    inc_bill = int(input("Please enter the expenditure:"))
    freq_bill=int(input("Please enter the frequency of the expenditure per month:")) #no:of times the expenditure spent in a month
    complete_bill= inc_bill*freq_bill 
    overallbudget = overallbudget - complete_bill
    check_budget(overallbudget)
    return (overallbudget)


#Subraction of expenses 
#Here we enter the amount that was not spent from overall budget
def subtract_expenses(overallbudget):    
    dec_bill = int(input("Please enter the amount to be subtracted from expenses:"))
    freq1_bill = int(input("Please enter the frequency of the amount per month:"))
    bill=dec_bill*freq1_bill
    overallbudget = overallbudget + bill      #we add since the amount was not spent 
    check_budget(overallbudget)
    return (overallbudget)


#Additional Revenue/Income
def add_revenue(overallbudget):
    extra_income = int(input("Please enter the addtional income:"))
    overallbudget = overallbudget + extra_income
    check_budget(overallbudget)
    return (overallbudget)

#Subtraction of revenue
def subtract_revenue(overallbudget):
    loss = int(input("Please enter the amount to be subtracted from income:"))
    overallbudget = overallbudget - loss 
    check_budget(overallbudget)
    return (overallbudget)
    
# To check whether the amount has exceeded the limit   
def check_budget(overallbudget):
    if (overallbudget >= 0):
        print("The overall remaining budget is:", overallbudget)
    else:
        print("You have exceeded your monthly budget!\n\nRe-evaluate the expenses and balance your budget.")
       
        print("Your current Balance is:", overallbudget)
  
 #to invoke main function
main()  
```
